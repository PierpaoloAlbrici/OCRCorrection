```python
import pymongo
import string
from tqdm.notebook import tqdm

client = pymongo.MongoClient('mongodb://localhost:27017')
db = client['vatican']
dataset = db['phrasesDataset']
wordsDataset = db['wordsDataset']

docs = list(dataset.find())
```

```python
#wordsDataset.drop()
wordsDataset.count()
```

```python
#la lista dei caratteri è da rifare

chars = string.ascii_lowercase + string.whitespace + '#'
CHAR_INDEX = dict((c, i) for i, c in enumerate(chars))
chars

def preprocess(s):
    out = ""
    for c in s.lower():
        if c in CHAR_INDEX.keys():
            out += c
        else:
            out += '#'
    return "".join(out)
```

```python
word = dict()

for i, d in tqdm(enumerate(docs)):
    for j, s in enumerate(d['sentences']):
        for k, w in enumerate(s[0]):
            word.setdefault(s[1][k], []).append(preprocess(w))            
            
for wk in word.keys():
    item = dict(Counter(word.get(wk)))
    wordsDataset.insert_one({'word': preprocess(w), 'occurrences': item})
    
```
